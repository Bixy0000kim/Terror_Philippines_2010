```python
from selenium import webdriver
from bs4 import BeautifulSoup 
```

```python
browser = webdriver.Edge()
```

```python
url = 'https://en.wikipedia.org/wiki/Terrorism_in_the_Philippines#List_of_terrorist_incidents_in_the_Philippines'
```

```python
browser.get(url)
```

```python
html = browser.page_source
html
```

```python
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
soup.select('table.wikitable.sortable.jquery-tablesorter')[14].find_all('thead')[0].find_all('th')[0].text.strip()
```

```python
table2 = soup.select('table.wikitable.sortable.jquery-tablesorter')[14]
```

```python
table2.find_all('th')[0].text.strip()
```

```python
Top = [i.text.strip() for i in table2.find_all('th')]
Top
```

```python
import pandas as pd
from tqdm import tqdm
```

```python
def Terror2010():
    Top = [i.text.strip() for i in table2.find_all('th')]
    data = []
    for r in table2.tbody.find_all('tr'):
        data.append([i.text.strip() for i in r.find_all('td')])
    df = pd.DataFrame(data)
    df.columns = Top
    return df
```

```python
Terror2010()
```

```python
df20 = Terror2010()
```

```python
df20.to_csv('Terror_Philippines_2010.csv')
```
